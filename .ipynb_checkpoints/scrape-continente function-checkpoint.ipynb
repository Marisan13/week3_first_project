```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup

def scrape_continente(ingredient):
    #url = f"https://www.continente.pt/pesquisa/?q={ingredient}&start=0&srule=Continente&pmin=0.01"
    url = f"https://www.continente.pt/pesquisa/?q={ingredient}&start=0&srule=Continente&pmin=0.01&prefn1=brand&prefv1=Continente%7CContinente%20Bio%7CContinente%20Equil%C3%ADbrio%7CContinente%20Infantil"
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    
    product_dict = {}
    index = 0
    
    products = soup.find_all("div", attrs = {"class":"product"})
    
    for product in products:

        # Scrape brand name  
        description_brand = product.find_all("p", attrs = {"class":"pwc-tile--brand col-tile--brand"})
        if description_brand:
            product_brand = description_brand[0].get_text().strip()
        # Scrape the name of the product            
        description_name = product.find_all("div", attrs = {"class":"ct-pdp-link col-pdp-link"})
        if description_name:
            product_name = description_name[0].get_text().strip()
        # Scrape the price of the product        
        description_price = product.find_all("span", attrs = {"class":"value"})
        if description_price:
            product_price = description_price[0].get_text().strip()
            # Define a regular expression pattern to match the numeric part
            # price_pattern = r'€([\d,]+)'
            # Use the re.search function to find the first match
            # match = re.search(price_pattern, product_price)
            # if match:
                #product_price = match.group(1)  # Group 1 contains the numeric part
                #product_price = product_price.replace(',', '.')  # Replace comma with a dot for a valid numeric format
        # Scrape the quantity of the product        
        description_quantity = product.find_all("p", attrs = {"class":"pwc-tile--quantity col-tile--quantity"})
        if description_quantity:
            product_quantity = description_quantity[0].get_text().strip()

        product_dict[index] = {"category": ingredient,
                               "brand":product_brand, 
                               "name":product_name,
                               "price":product_price,
                               "quantity":product_quantity}
        index +=1
    
    return pd.DataFrame.from_dict(product_dict, orient="index")[:1]

ingredients = ["Arroz", "Feijão enlatado","Macarrão", "Leite","Ovos","Pão","Frango","Carne moída","Batatas","Cebola","Tomate","Alface","Banana","Maçã","Laranja","Azeite","Sal","Açúcar","Café","Água mineral","Atum","Peixe","Pasta", "Beans"]
dataframes = []

for ingredient in ingredients:
    df = scrape_continente(ingredient)
    dataframes.append(df)

# Concatenate the dataframes
ingredient_df = pd.concat(dataframes, ignore_index=True)

# Print the concatenated dataframe
display(ingredient_df)
```

```python
# Function to regex the price column values of the ingredient table. 

def extract_first_number(intro_string):
    match = re.search(r'[€$]?\d+\,\d{2}', intro_string)
    if match:
        price_str = match.group()
        price = float(price_str.replace('€', '').replace('$', '').replace(',','.'))
        return price
    else:
        return None
```
